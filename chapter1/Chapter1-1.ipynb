```python
from transformers import pipeline
```

## Sentiment analysis

```python
text_classification_pipeline = pipeline(
    model="llm-book/bert-base-japanese-v3-marc_ja"
)

# 適当な文を入力
text = input("文章を入力: ")

# 入力された分の極性を予測
print(text_classification_pipeline(text)[0])  # result: positive | nevative
```

## Natural language inference

```python
nli_pipeline = pipeline(model="llm-book/bert-base-japanese-v3-jnli")

# 文章を入力
text1 = input("一つ目の文章を入力: ")
text2 = input("二つ目の文章を入力: ")

# 入力したそれぞれの文章の論理関係を予測
print(nli_pipeline({"text": text1, "text_pair": text2})) # result: entailment | contradiction | neutral
```

## Semantic textual similarity

```python
sts_pipeline  = pipeline(
    model="llm-book/bert-base-japanese-v3-jsts",
    function_to_apply="none"
)

text1 = input("1つ目の文章を入力: ")
text2 = input("2つ目の文章を入力: ")

# 入力した文が似ているかどうかを予測する
result = sts_pipeline({"text": text1, "text_pair": text2})
print(result["score"]) # result: 0-5 (数値が大きいほど文が似ている)
```

## Sentence embed-ding

```python
from torch.nn.functional import cosine_similarity

sim_enc_pipeline = pipeline(
    model="llm-book/bert-base-japanese-v3-unsup-simcse-jawiki",
    task="feature-extraction",
)

text1 = input("1つ目の文章を入力: ")
text2 = input("2つ目の文章を入力: ")

# それぞれの文のベクトルを獲得する
text1_emb = sim_enc_pipeline(text1, return_tensors=True)[0][0]
text2_emb = sim_enc_pipeline(text2, return_tensors=True)[0][0]

# それぞれの文の類似度を計算する
sim_par_score = cosine_similarity(text1_emb, text2_emb, dim=0)
print(f"result: {sim_par_score.item()}") # result -1 - 1 (正の値に近いほど文が類似している)
```

```python
# 出力チェック(遊び)
test = sim_enc_pipeline(text1, return_tensors=True)
print(test)
```

## Named entity recognition

```python
from pprint import pprint

ner_pipeline = pipeline(
    model="llm-book/bert-base-japanese-v3-ner-wikipedia-dataset",
    aggregation_strategy="simple"
)

text = input("文章を入力")

# 文中の固有表現を抽出
# result: 'end': 終了位置, 'entity_group': 固有表現の種類, 'score': 予想スコア, 'start': 固有表現の開始位置, 'word': 抽出したワード
pprint(ner_pipeline(text))
```

## Sumarization generation

```python
text2text_pipeline = pipeline(
    model="llm-book/t5-base-long-livedoor-news-corpus"
)

text = """
Geminiは、テキスト／画像／音声／数値など複数の種類のデータ（モダリティ）を処理できるマルチモーダルAI（人工知能）モデル。
Googleは、Gemini 1.5の初期テスト用モデルとして「Gemini 1.5 Pro」の提供を開始している。
Gemini 1.5 Proは、12万8000トークンのコンテキストウィンドウを持つ中規模のマルチモーダルモデルだ。
"""

# 入力した文の要約を生成
print(text2text_pipeline(text)[0]["generated_text"])
```

```python
# 出力チェック (遊び)
print(text2text_pipeline(text)) # list
print(text2text_pipeline(text)[0]) # dict
```
