```python
from transformers import pipeline
```

```python
# GPTの日本語モデルを使って後続するトークンを予測する
generator = pipeline(
    "text-generation", model="abeja/gpt2-large-japanese"
)

inputs = input("文章を入力: ")

# 後続するトークンのテキストを生成する
outputs = generator(inputs)
print(outputs[0]["generated_text"])
```

```python
# 出力確認
print(outputs) # list
print(*outputs) # dict
```
