## BERTモデルを使用して、マスクされたトークンを予測する

```python
from transformers import pipeline
import pandas as pd

# マスクされたトークンを予測するのpipelineの作成
# fill-mask: トークンの穴埋めを行うために必要
fill_mask = pipeline(
    "fill-mask", model="cl-tohoku/bert-base-japanese-v3"
)

masked_text = "日本の首都は[MASK]である"
# マスクされた部分のトークンを予測する
output = fill_mask(masked_text)

# 上位3つのトークンを表示
display(pd.DataFrame(output[:3]))
```

```python
# 評判の分析を行う
masked_text = "今日はいい試合だった。この試合は私にとって[MASK]だった"
output = fill_mask(masked_text)
display(pd.DataFrame(output[:3]))
```
